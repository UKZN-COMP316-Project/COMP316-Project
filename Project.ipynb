```python
import pandas as pd
```

```python
data=pd.read_csv('/content/goodreads_project_data.csv',encoding_errors='ignore',delimiter=';')
```

Data Cleaning

```python
#Creating a copy of the original data so that we do not mess up our original file
books=data.copy()
# books.head(3)
```

```python
#dropping all the columns we don't need
books=books.drop(['URL','Num_Ratings','Avg_Rating','Author'], axis=1)
```

```python
books.sample(5)
```

```python
#removing rows that have null values in the description column for data consistency
books= books.dropna(subset='Description')
books.info()
```
